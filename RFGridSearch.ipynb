```python
Necessary libraries are imported
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load dataset
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
#Preprocessing the data
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
#Splitting the input  and output variables
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
 dep=dataset['charges']
```

```python
#importing the Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
#parameter values to be tuned
param_grid = {'criterion':['squared_error','friedman_mse'], 'max_features': [1.0,'sqrt','log2'], 'n_estimators':[10,100]}
#fitting the grid with the dataset
grid = GridSearchCV(RandomForestRegressor(), param_grid, refit = True, verbose=3,n_jobs=-1)
grid.fit(indep, dep)
```

```python
#getting the results
re=grid.cv_results_
```

```python
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
#Tabularized form
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
age_input=float(input("Age:")) 
bmi_input=float(input("BMI:")) 
children_input=float(input("Children:")) 
sex_male_input=int(input("Sex Male 0 or 1:")) 
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
#prediction
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]]) 
```

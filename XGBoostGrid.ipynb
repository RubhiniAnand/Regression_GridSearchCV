```python
#Importing the libraries needed
import pandas as pd
```

```python
#Loading the dataset
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#dataset is preprocessed
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
#declaring and defining the independent and dependent variables
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dep=dataset[['charges']]
```

```python
#GridSearchCV is imported from the sklearn library
#XGBRegressor module is imported from the sklearn
from sklearn.model_selection import GridSearchCV
from xgboost.sklearn import XGBRegressor
```

```python
#creating a dictionary that contains the hyperparameters to be tuned
param_grid = {'learning_rate': [0.05,0.10,0.15],'max_depth':[3,4,5,6,8],'min_child_weight':[1,3,5,7],}
```

```python
#call the parameters defined in the GridSearchCV()
#fitting the model
grid = GridSearchCV(XGBRegressor(), param_grid, refit = True, verbose= 0,n_jobs=-1)
grid.fit(indep,dep)
```

```python
#prints all the results and the best parameter
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
#the results will be displayed in a tabularized form
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
#predicting the output
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```

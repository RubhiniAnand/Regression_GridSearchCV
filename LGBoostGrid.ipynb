```python
#Importing the libraries needed
import pandas as pd
```

```python
#Loading the dataset
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#dataset is preprocessed
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
#declaring and defining the independent and dependent variables
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dep=dataset[['charges']]
```

```python
#importing the GridearchCV
from sklearn.model_selection import GridSearchCV
```

```python
#LGBMRegressor is imported
from lightgbm import LGBMRegressor
#Hyperparameters to be tuned
param_grid = {'learning_rate': [0.05,0.1,0.2],'n_estimators':[50,100,150],'num_leaves':[5,20,31]}
```

```python
#fits the model
grid = GridSearchCV(LGBMRegressor(), param_grid, refit = True, verbose= 0,n_jobs=-1)
grid.fit(indep,dep)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```

```python
#libraries
import pandas as pd
```

```python
#raw data
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#One Hot Encoding technique to prepare the data
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
#Defining the input and output variables
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dep=dataset[['charges']]
```

```python
#import GridSearchCV module and AdaBoostRegressor to get the predicted output
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostRegressor
```

```python
#parameters to be tuned 
param_grid = {'n_estimators':[50,100,200,500],
'learning_rate': [0.0001,0.001,0.01,0.1,1.0]}
```

```python
#fitting the model
grid = GridSearchCV(AdaBoostRegressor(), param_grid, refit = True, verbose= 3,n_jobs=-1)
grid.fit(indep,dep)
```

```python
#results stored in 're'variable
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
#tabularized form
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children:"))
sex_male_input=int(input("Sex Male 0 or 1:"))
smoker_yes_input=int(input("Smoker Yes 0 or 1:"))
```

```python
#pradicting
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
